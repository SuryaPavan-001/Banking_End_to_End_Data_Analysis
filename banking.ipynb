```python
bins = [0,100000,300000,float("inf")]
labels = ["low","med","high"]

df["Income Band"] = pd.cut(df["Estimated Income"],bins =bins , labels= labels, right=False)
```

```python
df["Income Band"].value_counts().plot( kind = "bar")
```

```python
 #Examine the distribution of unique categories in categorical columns (univ analysis)

categorical_cols = df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Risk Weighting","Properties Owned","Income Band"]].columns


for col in categorical_cols:
    print(f"value counts for ' {col}':")
    display(df[col].value_counts())
```

## UNIVARIATE ANALYSIS


```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Risk Weighting","Properties Owned","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

## BIVARIATE ANALYSIS

```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Risk Weighting","Properties Owned","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="GenderId")
```

```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Risk Weighting","Properties Owned","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="Nationality")
```

```python
#HISTOGRAM of value counts for different Occupations

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(10,5))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

## NUMERICAL ANALYSIS

```python
df.columns
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings', "Credit Card Balance", "Bank Loans",'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
       'Foreign Currency Account', 'Business Lending' ]

#univariate analysis and visualization

plt.figure(figsize=(20,21))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde = True)
    plt.title(col)
plt.show()
```

## HEATMAP


```python
numerical_cols = ['Estimated Income','Superannuation Savings', "Credit Card Balance", "Bank Loans",'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
       'Foreign Currency Account', 'Business Lending' ]

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest' , fmt= ".2f")
plt.title("Correlation Matrix")
plt.show()
```
